### Clustering

### Data Preparation

let's transform categorical features into numeric ones, as well as scaling the data due to distance being a key factor in clustering.

```python
df.head()
```

```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

### Creating and Fitting a KMeans Model

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
# IMPORTANT NOTE: YOUR 0s and 1s may be opposite of ours,
# makes sense, the number values are not significant!
cluster_labels
```

```python
len(scaled_X)
```

```python
len(cluster_labels)
```

```python
X['Cluster'] = cluster_labels
```

```python
sns.heatmap(X.corr())
```

```python
X.corr()['Cluster']
```

```python
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

### Choosing K Value

```python
ssd = []

for k in range(2,10):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(scaled_X)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,10),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

Analyzing SSE Reduction
